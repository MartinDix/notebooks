```python

import xarray as xr
import iris, iris.plot as iplt, iris.quickplot as qplt
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
orog = iris.load_cube('/g/data/access/umdir/ancil/atmos/GC5/n96e/orca1/orography/gmted_ramp2/v1/qrparm.orog_mean')

fig = plt.figure()
axes = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
axes.coastlines(resolution='110m')
p = iplt.pcolormesh(orog)
axes.set_title('N96 GC5 orography')
plt.colorbar(p, orientation='horizontal')
```

```python
orog = iris.load_cube('/g/data/access/umdir/ancil/atmos/n1280e/orography/globe30/v2/qrparm.orog', 'surface_altitude')
fig = plt.figure()
axes = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
axes.coastlines(resolution='110m')
p = iplt.pcolormesh(orog)
axes.set_title('N1280 orography')
plt.colorbar(p, orientation='horizontal')
```

```python
fig = plt.figure()
axes = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
axes.coastlines(resolution='110m')
p = iplt.pcolormesh(orog, vmin=0,vmax=1500)
axes.set_extent((110,155,-45,-10))
axes.set_title('N1280 orography')
plt.colorbar(p, orientation='horizontal', extend='both')
```
